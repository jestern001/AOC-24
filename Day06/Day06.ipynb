# Setup

```python
from pathlib import Path
```

```python
TEST = False

# get data path
file_name = 'test-input.txt' if TEST else "input.txt"
file_path = Path(file_name)
if not file_path.exists():
    raise FileNotFoundError(file_path)

# get data
raw_data = file_path.read_text()
raw_rows = raw_data.split('\n')
rows = filter(None, raw_rows)
rows = map(list, rows)
rows = list(rows)
rows
```

# Part 1

Get x locations

```python
from math import sin, cos, degrees, radians, pi
```

```python
start = next((r, v) for r, row in enumerate(rows) for v, val in enumerate(row) if val == "^")
pos = start
dir = 0
start
```

```python
steps = 1
in_range = lambda pos: 0<=pos[0]<len(rows) and 0<=pos[1]<len(rows[0])
while pos:
    # assuming 0 is up
    for offset in [0, 90, 180, 270]:
        next_dir = dir+offset
        next_rad = radians(next_dir)
        next_move = (
            int(-1*cos(radians(next_dir))),
            int(sin(radians(next_dir)))
        )
        next_pos = (
            pos[0]+next_move[0],
            pos[1]+next_move[1]
        )
        if not in_range(next_pos):
            pos = None
            break
        next_space = rows[next_pos[0]][next_pos[1]]
        next_space_free = next_space != "#"
        if next_space_free:
            steps += 1
            pos = next_pos
            dir = degrees(next_rad) % 360
            print(f"{pos=}, {dir=}")
            break
steps += 1
print(pos, dir, steps)
```
