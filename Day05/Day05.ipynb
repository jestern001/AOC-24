# Setup

```python
from pathlib import Path
```

```python
TEST = False

# get data path
file_name = 'test-input.txt' if TEST else "input.txt"
file_path = Path(file_name)
if not file_path.exists():
    raise FileNotFoundError(file_path)

# get data
raw_data = file_path.read_text()
raw_rows = raw_data.split('\n\n')
```

```python

orders = raw_rows[0].split('\n')
orders = [row.split('|') for row in orders]
orders = [list(map(int, row)) for row in orders]
orders[:10]
```

```python

updates = raw_rows[1].split('\n')
updates = [row.split(',') for row in updates if row]
updates = [list(map(int, row)) for row in updates]
updates[0][:10]
```

# Part 1

```python
valid_rows = []
for u in updates:
    applicable_orders = [o for o in orders if o[0] in u and o[1] in u]
    if all(u.index(pre) < u.index(post) for pre, post in applicable_orders):
        valid_rows.append(u)
valid_rows[0][:5]
```

```python
mids = [row[len(row)//2] for row in valid_rows]
mids[:5]
    
```

```python
result = sum(mids)
```

```python
if TEST:
    assert result == 143
result
```

# Part 2

My solution for part 2 was too complex, I didn't have the time to refine it.  Posting as is.

```python
invalid_rows = [row for row in updates if row not in valid_rows]
invalid_rows
```

```python
from itertools import permutations


# this is a bit brute force, but it was quick and easy to write and I only have so much time
# there's probably a better solution that involves picking applicable orders that lowers complexity
fixed_rows = []
for r in invalid_rows:
    applicable_orders = [o for o in orders if o[0] in r and o[1] in r]
    updates_to_check = permutations(r)
    for u in updates_to_check:
        if all(u.index(pre) < u.index(post) for pre, post in applicable_orders):
            fixed_rows.append(u)
            break
fixed_rows    
```

```python
mids = [row[len(row)//2] for row in fixed_rows]
mids[:5]
```

```python
result = sum(mids)
```

```python
if TEST:
    assert result == 123
result
```
